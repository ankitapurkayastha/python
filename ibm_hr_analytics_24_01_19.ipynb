```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('attrition.csv')
data
```

```python
Attrition=data.groupby('Attrition')
Attrition.mean()
```

```python
data.describe()
```

```python
Attrition_count=data.groupby('Attrition').count()
plt.bar(Attrition_count.index.values,Attrition_count['RelationshipSatisfaction'])
plt.xlabel('Employee left the company')
plt.ylabel('Number of employees')
plt.show()
```

```python
from sklearn.cluster import KMeans
#filter data
Attrition_emp= data[['JobSatisfaction', 'PerformanceRating']][data.Attrition == 'Yes']
#create groups using kmeans
kmeans = KMeans(n_clusters = 3, random_state = 0).fit(Attrition_emp)
```

```python
#add new columns
Attrition_emp['label'] = kmeans.labels_
#draw plot
plt.scatter(Attrition_emp["JobSatisfaction"],Attrition_emp["PerformanceRating"], c=Attrition_emp["label"], cmap="Accent")
plt.xlabel("JobSatisfaction")
plt.ylabel("PerformanceRating")
plt.title("3 Clusters of employee who left")
plt.show()
```

```python
Attrition_emp= data[['MonthlyIncome', 'PerformanceRating']][data.Attrition == 'Yes']
#create groups using kmeans
kmeans = KMeans(n_clusters = 3, random_state = 0).fit(Attrition_emp)
Attrition_emp['label'] = kmeans.labels_
#draw plot
plt.scatter(Attrition_emp["MonthlyIncome"],Attrition_emp["PerformanceRating"], c=Attrition_emp["label"], cmap="Accent")
plt.xlabel("MonthlyIncome")
plt.ylabel("PerformanceRating")
plt.title("3 Clusters of employee who left")
plt.show()
```
